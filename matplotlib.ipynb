Grfico de linha

```python
# Dados
x = linspace(0,5,10)
y = x ** 2

# Criar a figura

fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0,0,0.8,0.8])

# Criar o plot
axes.plot(x,y,'r')

# Labels e ttulo
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Grfico de Linha')
```

```python
# Dados
x = linspace(0,5,10)
y = x ** 2

# Criar a figura

fig = plt.figure()

# Define a escala dos eixos
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # eixos da figura secundria

# Figura principal
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# Figura secundria
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundria')

```

```python
x = linspace(0,5,10)
y = x ** 2


# Divide a rea de plotagem em dois subplots
fig, axes = plt.subplots(nrows=1, ncols=2) 

# Loop pelo eixos para criar cada plot
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Ttulo')

fig.tight_layout()
```

```python
x = linspace(0,5,10)
y = x ** 2


# Criando os subplots
fig, axes = plt.subplots(1,2,figsize=(10,4))

# Criando o plot1

axes[0].plot(x,x**2,x,exp(x))
axes[0].set_title('Escala Padro')

# Criando o plot2

axes[1].plot(x,x**2,x,exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)')
```

```python
x = linspace(0,5,10)
y = x ** 2


# Criando os subplots
fig, axes = plt.subplots(1,2,figsize=(10,3))

# Grid padro

axes[0].plot(x,x**2,x,x**3,lw=2)
axes[0].grid(True)

# Grid customizado

axes[1].plot(x,x**2,x,x**3,lw=2)
axes[1].grid(color='b', alpha= 0.7, linestyle='dashed', linewidth=0.8)
```

```python
xx = np.linspace(-0.75,1.,100)
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color= 'black')
axes[0].set_title('scatter')

axes[1].step(n,n**2,lw=2,color='blue')
axes[1].set_title('step')

axes[2].bar(n,n**2,align='center',width=0.5,alpha=0.5, color='magenta')
axes[2].set_title('bar')

axes[3].fill_between(x,x**2,x**3,alpha=0.5, color='green')
axes[3].set_title('fill_between')
```

Histograma

```python
n = np.random.randn(100000)

fig, axes = plt.subplots(1,2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title('Histograma Padro')
axes[0].set_xlim(min(n),max(n))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title('Histograma Cumulativo')
axes[1].set_xlim(min(n),max(n))
```
